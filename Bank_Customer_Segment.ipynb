```python
# L√ÄM S·∫†CH D·ªÆ LI·ªÜU C∆† B·∫¢N
# ============================================================================
# Chu·∫©n h√≥a t√™n c·ªôt
users = users.rename(columns={"id": "client_id"})
users.columns = users.columns.str.strip().str.lower()
transactions.columns = transactions.columns.str.strip().str.lower()
cards.columns = cards.columns.str.strip().str.lower()

# X·ª≠ l√Ω c·ªôt ti·ªÅn t·ªá
money_cols_tx = ['amount']
for col in money_cols_tx:
    if col in transactions.columns:
        transactions[col] = transactions[col].replace('[\$,]', '', regex=True).astype(float)

money_cols_cards = ['credit_limit']
for col in money_cols_cards:
    if col in cards.columns:
        cards[col] = cards[col].replace('[\$,]', '', regex=True).astype(float)

money_cols_users = ['yearly_income', 'total_debt', 'per_capita_income']
for col in money_cols_users:
    if col in users.columns:
        users[col] = users[col].replace('[\$,]', '', regex=True).astype(float)
```

```python
# T·ªîNG H·ª¢P D·ªÆ LI·ªÜU GIAO D·ªäCH (TRANSACTIONS)
tx_agg = transactions.groupby("client_id").agg({
    "amount": ["count", "sum", "max", "min"],
    "use_chip": lambda x: (x == "Chip Transaction").mean(),
    "errors": lambda x: (x.astype(str) != "NaN").mean()
}).reset_index()

tx_agg.columns = [
    "client_id", "transaction_count", "total_amount",
    "max_amount", "min_amount", "chip_transaction_ratio",
    "error_transaction_ratio"
]

tx_agg["avg_amount"] = tx_agg["total_amount"] / tx_agg["transaction_count"]

# N·∫øu c√≥ giao d·ªãch online th√¨ th√™m t·ª∑ l·ªá giao d·ªãch online
tx_agg["online_transaction_ratio"] = (
    transactions.groupby("client_id")
    .apply(lambda x: (x["use_chip"] == "Online Transaction").sum() / len(x))
    .reindex(tx_agg["client_id"], fill_value=0)
    .values
)

# ============================================================================
# T·ªîNG H·ª¢P D·ªÆ LI·ªÜU TH·∫∫ (CARDS)
card_agg = cards.groupby("client_id").agg({
    "id": "count",
    "credit_limit": "mean",
    "num_cards_issued": "sum",
    "card_on_dark_web": lambda x: (x.astype(str).str.lower().isin(["yes", "true", "1"])).any()
}).reset_index()

card_agg.columns = [
    "client_id", "num_cards", "avg_credit_limit",
    "total_cards_issued", "is_card_on_dark_web"
]

# ============================================================================
# G·ªòP D·ªÆ LI·ªÜU
df = (
    users
    .merge(tx_agg, on="client_id", how="left")
    .merge(card_agg, on="client_id", how="left")
)
```

```python
# Lo·∫°i b·ªè outliers
def remove_outliers_iqr(df, outlier_threshold=1.5):
    df_clean = df.copy()
    numeric_cols = df_clean.select_dtypes(include=["float64", "int64"]).columns

    # Kh·ªüi t·∫°o mask to√†n True
    mask = pd.Series([True] * len(df_clean), index=df_clean.index)

    for col in numeric_cols:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - outlier_threshold * IQR
        upper_bound = Q3 + outlier_threshold * IQR

        # Update mask
        mask &= (df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)

    return df_clean[mask]
print(f"üìä Tr∆∞·ªõc lo·∫°i b·ªè outliers: {df.shape}")
df_cleaned = remove_outliers_iqr(df, outlier_threshold=1.5)
print(f"üìä Sau lo·∫°i b·ªè outliers: {df_cleaned.shape}")
```

# PH√ÇN C·ª§M PH√ÇN C·∫§P

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# -------------------------
# Chu·∫©n b·ªã d·ªØ li·ªáu
features = ["total_debt", "avg_amount", "credit_score"]
_cluster = df_model[["client_id"] + features].copy()
num_features = ["total_debt", "avg_amount", "credit_score"]

# -------------------------
# C·∫•u h√¨nh th·ª≠ nghi·ªám
scalers = {
    "Standard": StandardScaler()
}

pca_components = 3
k_range = range(3, 11)

results = []

# -------------------------
# Loop qua t·ª´ng scaler
for scaler_name, scaler in scalers.items():
    X_scaled = scaler.fit_transform(_cluster[num_features])

    # PCA 3D
    pca = PCA(n_components=pca_components, random_state=42)
    X_pca = pca.fit_transform(X_scaled)

    # Th·ª≠ c√°c gi√° tr·ªã k
    for k in k_range:
        model = AgglomerativeClustering(n_clusters=k, linkage='ward')
        labels = model.fit_predict(X_pca)
        sil = silhouette_score(X_pca, labels)

        results.append({
            "scaler": scaler_name,
            "pca_comp": pca_components,
            "n_clusters": k,
            "silhouette": sil
        })

# -------------------------
# T·∫°o DataFrame k·∫øt qu·∫£
res_df = pd.DataFrame(results)

# -------------------------
# V·∫Ω bi·ªÉu ƒë·ªì silhouette theo k
plt.figure(figsize=(8,5))
for scaler_name in scalers.keys():
    subset = res_df[res_df["scaler"] == scaler_name]
    plt.plot(subset["n_clusters"], subset["silhouette"], marker='o', label=scaler_name)

plt.title("Silhouette Score theo s·ªë c·ª•m (Agglomerative)")
plt.xlabel("S·ªë c·ª•m (k)")
plt.ylabel("Silhouette Score")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------
# Hi·ªÉn th·ªã k·∫øt qu·∫£ chi ti·∫øt
display(res_df.sort_values(by=["scaler", "silhouette"], ascending=[True, False]).reset_index(drop=True))
```

```python
# C·∫•u h√¨nh m√¥ h√¨nh t·ªëi ∆∞u
best_scaler = StandardScaler()
best_pca_n = 3
best_n_clusters = 5

# ==========================
# Scale d·ªØ li·ªáu v√† gi·∫£m chi·ªÅu PCA
num_features = ["total_debt", "avg_amount", "credit_score"]
X_scaled = best_scaler.fit_transform(_cluster[num_features])

pca = PCA(n_components=best_pca_n, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# ==========================
# Ph√¢n c·ª•m b·∫±ng Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=best_n_clusters, linkage='ward')
labels_agg = agg.fit_predict(X_pca)
_cluster["Cluster"] = labels_agg

sil_agg = silhouette_score(X_pca, labels_agg)
print(f"Agglomerative silhouette (k={best_n_clusters}) = {sil_agg:.3f}")

# ==========================
# Visualize 3D PCA
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    X_pca[:,0], X_pca[:,1], X_pca[:,2],
    c=labels_agg, cmap='viridis', s=35, alpha=0.8
)
ax.set_title(f'Ph√¢n c·ª•m kh√°ch h√†ng (Agglomerative ‚Äî k = {best_n_clusters}\nSilhouette={sil_agg:.3f}')
ax.set_xlabel("PCA Component 1")
ax.set_ylabel("PCA Component 2")
ax.set_zlabel("PCA Component 3")
plt.show()

# ==========================
# Ph√¢n t√≠ch ƒë·∫∑c tr∆∞ng trung b√¨nh t·ª´ng c·ª•m
summary = (
    _cluster.groupby("Cluster")[num_features]
    .mean()
    .round(2)
)
print("ƒê·∫∑c tr∆∞ng trung b√¨nh c·ªßa t·ª´ng c·ª•m:")
display(summary)
```

```python
# Ph√¢n ph·ªëi s·ªë l∆∞·ª£ng kh√°ch h√†ng theo c·ª•m
cluster_counts = _cluster["Cluster"].value_counts().sort_index()

print("üìä S·ªë l∆∞·ª£ng kh√°ch h√†ng theo t·ª´ng c·ª•m:")
display(cluster_counts)

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(7,5))
bars = plt.bar(cluster_counts.index, cluster_counts.values,
               color=plt.cm.viridis(np.linspace(0,1,len(cluster_counts))))
plt.title("Ph√¢n ph·ªëi s·ªë l∆∞·ª£ng kh√°ch h√†ng theo c·ª•m (Agglomerative, k = 5)")
plt.xlabel("C·ª•m")
plt.ylabel("S·ªë l∆∞·ª£ng kh√°ch h√†ng")

# Ghi gi√° tr·ªã tr√™n ƒë·∫ßu c·ªôt
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
             f"{int(bar.get_height())}", ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Th√™m l·∫°i c·ªôt age_group ƒë·ªÉ ph√¢n t√≠ch s√¢u h∆°n
_cluster = _cluster.merge(
    df_model[["client_id", "age_group"]],
    on="client_id",
    how="left"
)

# Th·ªëng k√™ ph√¢n b·ªë nh√≥m tu·ªïi trong t·ª´ng c·ª•m
age_dist = (
    _cluster.groupby(["Cluster", "age_group"])
    .size()
    .unstack(fill_value=0)
)

print("üìä Ph√¢n b·ªë nh√≥m tu·ªïi trong t·ª´ng c·ª•m:")
display(age_dist)

# ==========================
# V·∫Ω bi·ªÉu ƒë·ªì stacked bar ƒë·ªÉ xem ph√¢n b·ªë
age_dist.plot(
    kind='bar',
    stacked=True,
    figsize=(8,5),
    colormap='viridis'
)
plt.title("Ph√¢n b·ªë nh√≥m tu·ªïi trong t·ª´ng c·ª•m kh√°ch h√†ng:")
plt.xlabel("C·ª•m kh√°ch h√†ng")
plt.ylabel("S·ªë l∆∞·ª£ng kh√°ch h√†ng")
plt.legend(title="Nh√≥m tu·ªïi")
plt.show()
```

```python
# Chia nh·ªè 5 c·ª•m theo nh√≥m tu·ªïi
# T·∫°o c·ªôt c·ª•m con: k·∫øt h·ª£p Cluster v√† age_group
_cluster["SubCluster"] = _cluster["Cluster"].astype(str) + "_" + _cluster["age_group"]

# Xem ph√¢n b·ªë s·ªë l∆∞·ª£ng kh√°ch h√†ng trong t·ª´ng c·ª•m con
subcluster_counts = _cluster["SubCluster"].value_counts().sort_index()
print("üìä Ph√¢n b·ªë kh√°ch h√†ng theo c·ª•m con:")
display(subcluster_counts)

# ==========================
# T√≠nh ƒë·∫∑c tr∆∞ng trung b√¨nh cho t·ª´ng c·ª•m con
sub_summary = (
    _cluster
    .groupby("SubCluster")[["total_debt", "avg_amount", "credit_score"]]
    .mean()
    .round(2)
)

print("\nüìã ƒê·∫∑c tr∆∞ng trung b√¨nh c·ªßa t·ª´ng c·ª•m con:")
display(sub_summary)
```

# PH√ÇN C·ª§M THEO K-MEANS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score

# C·∫•u h√¨nh
num_features = ["total_debt", "avg_amount", "credit_score"]
pca_components = 3
k_range = range(3, 11)
scalers = {
    "Standard": StandardScaler()
}
kmeans_n_init = 20
random_state = 42

# Chu·∫©n b·ªã d·ªØ li·ªáu
_df = df_model.dropna(subset=num_features).copy()
X_raw = _df[num_features].values

# L∆∞u k·∫øt qu·∫£ cho KMeans
km_records = []

for scaler_name, scaler in scalers.items():
    Xs = scaler.fit_transform(X_raw)
    pca = PCA(n_components=pca_components, random_state=random_state)
    Xp = pca.fit_transform(Xs)
    for k in k_range:
        km = KMeans(n_clusters=k, n_init=kmeans_n_init, random_state=random_state)
        labels_km = km.fit_predict(Xp)
        # metrics
        if len(np.unique(labels_km)) > 1:
            sil = silhouette_score(Xp, labels_km)
            db = davies_bouldin_score(Xp, labels_km)
        else:
            sil = np.nan
            db = np.nan
        km_records.append({
            "method": "KMeans",
            "scaler": scaler_name,
            "k": k,
            "silhouette": sil,
            "db": db
        })

km_df = pd.DataFrame(km_records)

# Hi·ªÉn th·ªã c√°c k·∫øt qu·∫£ t·ªët nh·∫•t (theo ch·ªâ s·ªë silhouette)
print("Top KMeans results (by silhouette):")
display(km_df.sort_values("silhouette", ascending=False).head(10))

# V·∫Ω bi·ªÉu ƒë·ªì ƒë∆∞·ªùng bi·ªÉu di·ªÖn ch·ªâ s·ªë silhouette theo k cho t·ª´ng lo·∫°i scaler
plt.figure(figsize=(8,5))
for name in scalers.keys():
    sub = km_df[km_df["scaler"]==name].sort_values("k")
    plt.plot(sub["k"], sub["silhouette"], marker='o', label=name)
plt.xlabel("k"); plt.ylabel("Silhouette"); plt.title("KMeans: Silhouette by k and scaler"); plt.grid(True); plt.legend()
plt.show()

# Ch·ªçn c·∫•u h√¨nh KMeans t·ªët nh·∫•t (c√≥ silhouette cao nh·∫•t)
best_km_row = km_df.sort_values("silhouette", ascending=False).iloc[0]
best_km_scaler_name = best_km_row["scaler"]
best_k_km = int(best_km_row["k"])
best_km_sil = float(best_km_row["silhouette"])
print(f"Best KMeans: scaler={best_km_scaler_name}, k={best_k_km}, sil={best_km_sil:.4f}")

# Hu·∫•n luy·ªán m√¥ h√¨nh KMeans cu·ªëi c√πng v·ªõi scaler ƒë√£ ch·ªçn (v√† gi·ªØ PCA ƒë·ªÉ nh·∫•t qu√°n)
best_scaler = scalers[best_km_scaler_name]
Xs_best = best_scaler.fit_transform(X_raw)
pca_best = PCA(n_components=pca_components, random_state=random_state)
Xp_best = pca_best.fit_transform(Xs_best)

kmeans_final = KMeans(n_clusters=best_k_km, n_init=kmeans_n_init, random_state=random_state)
labels_km_final = kmeans_final.fit_predict(Xp_best)

# T√≠nh to√°n c√°c ch·ªâ s·ªë ƒë√°nh gi√°
sil_km_final = silhouette_score(Xp_best, labels_km_final)
db_km_final = davies_bouldin_score(Xp_best, labels_km_final)
print(f"\nFinal KMeans (scaler={best_km_scaler_name}, k={best_k_km}) -> Silhouette={sil_km_final:.4f}, DB={db_km_final:.4f}")

# G·∫Øn nh√£n c·ª•m KMeans v√†o b·∫£n sao c·ªßa dataframe g·ªëc
_df_km = _df.copy().reset_index(drop=True)
_df_km["Cluster_KMeans"] = labels_km_final

print("KMeans cluster counts:")
display(_df_km["Cluster_KMeans"].value_counts().sort_index())

# T√≥m t·∫Øt trung b√¨nh theo t·ª´ng c·ª•m KMeans
summary_km = _df_km.groupby("Cluster_KMeans")[num_features].mean().round(2)
print("\nKMeans - Summary per Cluster:")
display(summary_km)
```

Do ƒë√≥, KMeans ƒë∆∞·ª£c xem l√† m√¥ h√¨nh ph√π h·ª£p h∆°n cho b√†i to√°n ph√¢n kh√∫c kh√°ch h√†ng trong b·ªëi c·∫£nh n√†y.

# CLASSIFICATION

```python
# Chu·∫©n b·ªã d·ªØ li·ªáu t·ª´ KMeans
# =====================================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

# Gi·∫£ s·ª≠ _df_km ƒë√£ c√≥ c√°c c·ªôt:
# ['client_id', 'total_debt', 'avg_amount', 'credit_score', 'age_group', 'Cluster_KMeans', 'SubCluster_KM']

# M√£ h√≥a nh√≥m tu·ªïi
age_mapping = {"Young": 0, "Middle-aged": 1, "Senior": 2, "Unknown": -1}
_df_km["age_group_num"] = _df_km["age_group"].map(age_mapping)

# Ch·ªçn bi·∫øn ƒë·∫∑c tr∆∞ng v√† nh√£n m·ª•c ti√™u
features = ["total_debt", "avg_amount", "credit_score", "age_group_num"]
target = "SubCluster_KM"

X = _df_km[features]
y = _df_km[target]

# M√£ h√≥a nh√£n SubCluster
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Chia d·ªØ li·ªáu train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)
```

```python
# Chu·∫©n h√≥a d·ªØ li·ªáu
# =====================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Khai b√°o c√°c m√¥ h√¨nh classification
# =====================================
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42)
}
```

```python
# Hu·∫•n luy·ªán v√† ƒë√°nh gi√° c√°c m√¥ h√¨nh
results = []

for name, model in models.items():
    # ch·ªçn d·ªØ li·ªáu scale n·∫øu c·∫ßn
    if name in ["Logistic Regression", "SVM"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

    # t√≠nh c√°c ch·ªâ s·ªë ƒë√°nh gi√°
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    results.append({
        "M√¥ h√¨nh": name,
        "Accuracy": round(acc, 3),
        "Precision": round(prec, 3),
        "Recall": round(rec, 3),
        "F1-score": round(f1, 3)
    })

# =====================================
# T·ªïng h·ª£p k·∫øt qu·∫£
result_df = pd.DataFrame(results).sort_values(by="F1-score", ascending=False)

print("\n========== So s√°nh c√°c m√¥ h√¨nh Classification ==========\n")
print(result_df.to_string(index=False, justify="center"))
```

```python
# In chi ti·∫øt m√¥ h√¨nh t·ªët nh·∫•t
best_model_name = result_df.iloc[0]['M√¥ h√¨nh']
best_model = models[best_model_name]

print(f"üîπ M√¥ h√¨nh t·ªët nh·∫•t: {best_model_name}\n")

if best_model_name in ["Logistic Regression", "SVM"]:
    y_pred_best = best_model.predict(X_test_scaled)
else:
    y_pred_best = best_model.predict(X_test)

print("üìä B√°o c√°o chi ti·∫øt:")
print(classification_report(y_test, y_pred_best, digits=3))
```

```python
# =====================================
# üíæ Hi·ªÉn th·ªã v√† l∆∞u k·∫øt qu·∫£ d·ª± ƒëo√°n kh√°ch h√†ng th·∫≠t
# =====================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

if "current_age" in _df_km.columns:
    _df_km["age"] = _df_km["current_age"]
else:
    def simulate_age(row):
        if row["age_group"] == "Young":
            return np.random.randint(18, 40)
        elif row["age_group"] == "Middle-aged":
            return np.random.randint(40, 65)
        else:
            return np.random.randint(65, 90)
    _df_km["age"] = _df_km.apply(simulate_age, axis=1)

# Gi·∫£i m√£ nh√£n th·ª±c t·∫ø v√† d·ª± ƒëo√°n
y_test_label = label_encoder.inverse_transform(y_test)
y_pred_label = label_encoder.inverse_transform(y_pred_best)

# T·∫°o DataFrame k·∫øt qu·∫£ test
export_test = X_test.copy().reset_index(drop=True)
export_test["Th·ª±c t·∫ø"] = y_test_label
export_test["D·ª± ƒëo√°n"] = y_pred_label

# Lo·∫°i b·ªè c·ªôt age_group_num n·∫øu c√≥
if "age_group_num" in export_test.columns:
    export_test = export_test.drop(columns=["age_group_num"], errors="ignore")

# G·∫Øn l·∫°i tu·ªïi th·∫≠t t·ª´ _df_km
merge_cols = ["total_debt", "avg_amount", "credit_score"]
_df_km_unique = _df_km.drop_duplicates(subset=merge_cols)
export_test = export_test.merge(
    _df_km_unique[merge_cols + ["age"]],
    on=merge_cols,
    how="left"
)

# Th√™m c·ªôt ki·ªÉm tra ƒë√∫ng/sai
export_test["Correct"] = export_test["Th·ª±c t·∫ø"] == export_test["D·ª± ƒëo√°n"]

# S·∫Øp x·∫øp l·∫°i th·ª© t·ª± c·ªôt
cols_order = ["age", "total_debt", "avg_amount", "credit_score",
              "Th·ª±c t·∫ø", "D·ª± ƒëo√°n", "Correct"]
export_test = export_test[cols_order]

# Th·ªëng k√™ nhanh & in m·∫´u
total = len(export_test)
n_correct = export_test["Correct"].sum()
print(f"\nüìä T·ªïng m·∫´u test: {total} ‚Äî D·ª± ƒëo√°n ƒë√∫ng: {n_correct} m·∫´u ({(n_correct/total)*100:.2f}%)\n")

print("‚úÖ M·ªôt s·ªë d·ª± ƒëo√°n m·∫´u:")
display(export_test.loc[export_test["Correct"]].head(10))

# Xu·∫•t file CSV ƒë·ªÉ ki·ªÉm tra ngo√†i
export_test.to_csv("test_predictions_KMeans.csv", index=False, encoding="utf-8-sig")
print("üíæ ƒê√£ l∆∞u file: test_predictions_KMeans.csv")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# Ma tr·∫≠n nh·∫ßm l·∫´n (decoded labels)
cm = confusion_matrix(y_test, y_pred_best)
labels_decoded = label_encoder.inverse_transform(np.unique(np.concatenate([y_test, y_pred_best])))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_encoder.classes_)
fig, ax = plt.subplots(figsize=(8,6))
disp.plot(ax=ax, cmap="Blues", xticks_rotation=90)
plt.title(f"Confusion Matrix - {best_model_name}")
plt.show()
```
